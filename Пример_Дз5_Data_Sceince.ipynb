Создание бинарной целевой переменной:

```python
threshold = data['Customer Lifetime Value'].median()
data['LTV_binary'] = (data['Customer Lifetime Value'] > threshold).astype(int)
#Бинарная переменная LTV_binary была создана на основе медианы LTV.
```

Обучение моделей и подбор гиперпараметров:

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Список категориальных признаков
categorical_features = ['State', 'Response', 'Coverage', 'Education', 'EmploymentStatus', 'Gender', 'Location Code', 'Marital Status', 'Policy Type', 'Policy', 'Renew Offer Type', 'Sales Channel', 'Vehicle Class', 'Vehicle Size']

# Преобразование категориальных признаков в числовые метки
label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Разделение данных на признаки и целевую переменную
X = data.drop(columns=['Customer Lifetime Value', 'Customer', 'Effective To Date', 'LTV_binary'])
y = data['LTV_binary']

# Стандартизация числовых признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Построение и оценка моделей:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate

# Создание модели логистической регрессии
log_reg = LogisticRegression(C=0.8, random_state=42, max_iter=500)

# Проведение кросс-валидации
cv_log_reg = cross_validate(log_reg, X_scaled, y, cv=5, scoring='roc_auc')

# Вывод результатов кросс-валидации
print("Результаты Логистической регрессии:", cv_log_reg['test_score'].mean())
```

```python
# Решающее дерево
from sklearn.tree import DecisionTreeClassifier


dec_tree = DecisionTreeClassifier(random_state = 42) # Можно найти более оптимальные гиперпараметры

cv_dec_tree = cross_validate(dec_tree, X_sc, y, cv = 5, scoring = 'roc_auc')

print("Результаты решающего дерева:", sum(cv_dec_tree['test_score']) / len(cv_dec_tree['test_score']))
```

```python
# Случайный лес
from sklearn.ensemble import RandomForestClassifier


ran_for = RandomForestClassifier(random_state = 42) # Можно найти более оптимальные гиперпараметры

cv_ran_for = cross_validate(ran_for, X_sc, y, cv = 5, scoring = 'roc_auc')

print("Результаты решающего дерева:", sum(cv_ran_for['test_score']) / len(cv_ran_for['test_score']))
```

Использованы три модели: логистическая регрессия, решающее дерево и случайный лес.
Проведена кросс-валидация с оценкой по метрике ROC-AUC.

На основе обученной модели случайного леса была проведена оценка важности признаков:

```python
# Обучение модели случайного леса на всех данных
ran_for.fit(X_scaled, y)

# Важность признаков
importances = ran_for.feature_importances_
feature_names = X.columns
feature_importance = pd.Series(importances, index=feature_names).sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=feature_importance, y=feature_importance.index)
plt.title('Важность признаков для случайного леса')
plt.show()
```

Анализ важности признаков показал, что наиболее важные переменные для классификации LTV включают месячную премию за автострахование, общее количество требований по страхованию, доход клиента и другие.

Модели классификации:

Логистическая регрессия показала относительно низкий результат, что может свидетельствовать о сложности разделения данных по линейной границе.

Решающее дерево и случайный лес показали высокую точность. Случайный лес оказался самым эффективным, вероятно из-за способности модели обрабатывать сложные и нелинейные зависимости.
